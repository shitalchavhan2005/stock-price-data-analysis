```python
# Data Visualization

#In this section, we visualize the stock market dataset using different charts.
```

```python
import matplotlib.pyplot as plt
```

```python
top_volume = df.groupby("symbol")["volume"].sum().sort_values(ascending=False).head(10)

top_volume.plot(kind="bar", figsize=(10,5))

plt.title("Top 10 Companies by Trading Volume")
plt.xlabel("Company Symbol")
plt.ylabel("Total Trading Volume")

plt.show()
```

```python
## Line Chart - Closing Price Trend
##This chart shows how the closing stock price changes over time.
```

```python
sample = df[df["symbol"] == df["symbol"].unique()[0]]

plt.figure(figsize=(12,6))

plt.plot(sample["date"], sample["close"])

plt.title("Closing Price Over Time")
plt.xlabel("Date")
plt.ylabel("Closing Price")

plt.xticks(rotation=45)

plt.show()
```

```python
## Histogram - Closing Price Distribution
##This chart shows how closing prices are distributed.
```

```python
plt.figure(figsize=(8,5))

plt.hist(df["close"], bins=30)

plt.title("Distribution of Closing Prices")
plt.xlabel("Closing Price")
plt.ylabel("Frequency")

plt.show()
```

```python
top_companies = df["symbol"].value_counts().head(5)

plt.figure(figsize=(7,7))

plt.pie(
    top_companies,
    labels=top_companies.index,
    autopct="%1.1f%%"
)

plt.title("Top 5 Companies by Number of Records")

plt.show()
```

```python
import seaborn as sns
```

```python
plt.figure(figsize=(8,6))

sns.heatmap(
    df.corr(numeric_only=True),
    annot=True,
    cmap="coolwarm"
)

plt.title("Correlation Matrix")

plt.show()
```
